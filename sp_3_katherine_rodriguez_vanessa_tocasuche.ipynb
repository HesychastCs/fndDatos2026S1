```python
# 1.3 Selecciona al menos tres variables numéricas para el análisis
select_features = ['alcohol' , 'total_phenols' , 'color_intensity']
```

```python
# 1.4.1 Aplicar el método de escalamiento

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Datos del dataset de vinos
valores_originales = df[select_features].values  # (178, 3)

# Aplicar StandardScaler
scaler = StandardScaler()
datos_escalados = scaler.fit_transform(valores_originales)

# Estadísticas
medias = np.mean(valores_originales, axis=0)
stds = np.std(valores_originales, axis=0)

colores_orig   = ['steelblue', 'royalblue', 'cornflowerblue']
colores_escal  = ['mediumseagreen', 'seagreen', 'darkgreen']

# Crear figura: 2 filas × 3 columnas
fig, axs = plt.subplots(2, 3, figsize=(14, 8))
fig.suptitle('StandardScaler — Antes y Después (Histograma)', fontsize=14, fontweight='bold')

for i, feat in enumerate(select_features):

    # ── Fila 1: Datos originales ──────────────────────────────────────
    axs[0, i].hist(valores_originales[:, i], bins=20,
                   color=colores_orig[i], alpha=0.7, edgecolor='white')
    axs[0, i].axvline(medias[i], color='red', linestyle='--',
                      linewidth=1.5, label=f'media={medias[i]:.2f}')
    axs[0, i].axvline(medias[i] + stds[i], color='gray', linestyle=':',
                      linewidth=1.2, label=f'±std={stds[i]:.2f}')
    axs[0, i].axvline(medias[i] - stds[i], color='gray', linestyle=':', linewidth=1.2)
    axs[0, i].set_title(f'Original: {feat}')
    axs[0, i].set_xlabel('Valor')
    axs[0, i].set_ylabel('Frecuencia')
    axs[0, i].legend(fontsize=8)
    axs[0, i].grid(True, alpha=0.3)

    # ── Fila 2: Datos escalados ───────────────────────────────────────
    axs[1, i].hist(datos_escalados[:, i], bins=20,
                   color=colores_escal[i], alpha=0.7, edgecolor='white')
    axs[1, i].axvline(0,  color='red',  linestyle='--',
                      linewidth=1.5, label='media = 0')
    axs[1, i].axvline(1,  color='gray', linestyle=':',
                      linewidth=1.2, label='±1 std')
    axs[1, i].axvline(-1, color='gray', linestyle=':', linewidth=1.2)
    axs[1, i].set_title(f'Escalado: {feat}')
    axs[1, i].set_xlabel('Valor z')
    axs[1, i].set_ylabel('Frecuencia')
    axs[1, i].legend(fontsize=8)
    axs[1, i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 1.4.3 Aplicar el método de escalamiento RobustScaler sin outliers

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler

# Datos del dataset de vinos
valores_originales = df[select_features].values  # (178, 3)

# Aplicar RobustScaler
scaler = RobustScaler()
datos_robustos = scaler.fit_transform(valores_originales)

# Estadísticas
medianas = np.median(valores_originales, axis=0)
q1 = np.percentile(valores_originales, 25, axis=0)
q3 = np.percentile(valores_originales, 75, axis=0)
iqrs = q3 - q1

# Crear figura
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('RobustScaler — Datos del Vino (sin outliers extremos)', fontsize=13, fontweight='bold')

# Gráfico 1: Datos ORIGINALES
axs[0].boxplot([valores_originales[:, i] for i in range(len(select_features))],
               labels=select_features, patch_artist=True,
               boxprops=dict(facecolor='steelblue', alpha=0.6))
axs[0].set_title('Antes — Datos Originales')
axs[0].set_ylabel('Valor')
axs[0].grid(True)

# Anotar escala de cada feature
for i, feat in enumerate(select_features):
    axs[0].text(i + 1, np.max(valores_originales[:, i]) * 1.02,
                f'max={np.max(valores_originales[:, i]):.1f}',
                ha='center', fontsize=8, color='steelblue')

# Gráfico 2: Datos ESCALADOS con RobustScaler
axs[1].boxplot([datos_robustos[:, i] for i in range(len(select_features))],
               labels=select_features, patch_artist=True,
               boxprops=dict(facecolor='orange', alpha=0.6))
axs[1].set_title('Después — RobustScaler aplicado')
axs[1].set_ylabel('Valor Escalado')
axs[1].axhline(y=0, color='red', linestyle='--', alpha=0.6, label='mediana = 0')
axs[1].axhline(y=1, color='gray', linestyle=':', alpha=0.5, label='±IQR')
axs[1].axhline(y=-1, color='gray', linestyle=':', alpha=0.5)
axs[1].legend(loc='upper right')
axs[1].grid(True)

# Anotar mediana e IQR por feature
for i, feat in enumerate(select_features):
    axs[1].text(0.05, 0.95 - i * 0.1,
                f'{feat}: med={medianas[i]:.2f}, IQR={iqrs[i]:.2f}',
                transform=axs[1].transAxes, fontsize=9,
                color=['darkgreen', 'darkred', 'steelblue'][i], va='top')

plt.tight_layout()
plt.show()
```

```python
# 1.4.3 Aplicar el método de escalamiento RobustScaler con outliers

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler

# Datos del dataset de vinos
valores_originales = df[select_features].values  # (178, 3)

# Aplicar RobustScaler
scaler = RobustScaler()
datos_robustos = scaler.fit_transform(valores_originales)

# Estadísticas
medianas = np.median(valores_originales, axis=0)
q1 = np.percentile(valores_originales, 25, axis=0)
q3 = np.percentile(valores_originales, 75, axis=0)
iqrs = q3 - q1

# Detectar outliers (regla 1.5 * IQR)
mascara_outliers = np.any(
    (valores_originales < (medianas - 1.5 * iqrs)) |
    (valores_originales > (medianas + 1.5 * iqrs)),
    axis=1
)
n_outliers = np.sum(mascara_outliers)

# Crear figura
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle(f'RobustScaler — Con outliers detectados ({n_outliers} filas)',
             fontsize=13, fontweight='bold')

# ── Gráfico 1: Datos ORIGINALES con outliers resaltados ──────────────
for i, feat in enumerate(select_features):
    # Puntos normales
    axs[0].scatter(
        np.where(~mascara_outliers)[0],
        valores_originales[~mascara_outliers, i],
        color='steelblue', alpha=0.4, s=20,
        label='Normal' if i == 0 else ''
    )
    # Outliers en rojo
    axs[0].scatter(
        np.where(mascara_outliers)[0],
        valores_originales[mascara_outliers, i],
        color='red', s=60, zorder=5,
        label='Outlier' if i == 0 else ''
    )

# Líneas de límite IQR por feature
colores_linea = ['steelblue', 'green', 'purple']
for i in range(len(select_features)):
    axs[0].axhline(medianas[i] + 1.5 * iqrs[i],
                   color=colores_linea[i], linestyle='--', alpha=0.4, linewidth=1)
    axs[0].axhline(medianas[i] - 1.5 * iqrs[i],
                   color=colores_linea[i], linestyle='--', alpha=0.4, linewidth=1)

axs[0].set_title('Antes — Datos Originales (outliers en rojo)')
axs[0].set_xlabel('Índice')
axs[0].set_ylabel('Valor')
axs[0].legend()
axs[0].grid(True)

# ── Gráfico 2: Datos ESCALADOS con outliers contenidos ───────────────
for i, feat in enumerate(select_features):
    # Puntos normales
    axs[1].scatter(
        np.where(~mascara_outliers)[0],
        datos_robustos[~mascara_outliers, i],
        color='orange', alpha=0.4, s=20,
        label='Normal' if i == 0 else ''
    )
    # Outliers en rojo (siguen siendo outliers pero contenidos)
    axs[1].scatter(
        np.where(mascara_outliers)[0],
        datos_robustos[mascara_outliers, i],
        color='red', s=60, zorder=5,
        label='Outlier (contenido)' if i == 0 else ''
    )

axs[1].axhline(y=0,  color='red',  linestyle='--', alpha=0.5, label='mediana = 0')
axs[1].axhline(y=1,  color='gray', linestyle=':',  alpha=0.5, label='±1 IQR')
axs[1].axhline(y=-1, color='gray', linestyle=':',  alpha=0.5)
axs[1].set_title('Después — RobustScaler (outliers contenidos)')
axs[1].set_xlabel('Índice')
axs[1].set_ylabel('Valor Escalado')
axs[1].legend(loc='upper right')
axs[1].grid(True)

# Anotar mediana e IQR
for i, feat in enumerate(select_features):
    axs[1].text(0.05, 0.95 - i * 0.1,
                f'{feat}: med={medianas[i]:.2f}, IQR={iqrs[i]:.2f}',
                transform=axs[1].transAxes, fontsize=9,
                color=['darkgreen', 'darkred', 'steelblue'][i], va='top')

plt.tight_layout()
plt.show()
```

```python
# 1.4.1 Aplicar el método de escalamiento Z-Score

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Datos del dataset de vinos
valores_originales = df[select_features].values

# Estandarización Z-score
datos_estandarizados = zscore(valores_originales, axis=0)

# Promedios por feature
medias_originales = valores_originales.mean(axis=0)
medias_zscore = datos_estandarizados.mean(axis=0)

indices = np.arange(1, len(select_features) + 1)

media = np.mean(valores_originales)
std = np.std(valores_originales)

# Crear figura
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Gráfico 1: Datos originales
axs[0].scatter(indices, medias_originales, color='steelblue', s=100, zorder=3)
axs[0].set_title('Datos Originales (promedio por feature)')
axs[0].set_xlabel('Feature')
axs[0].set_ylabel('Valor promedio')
axs[0].set_xticks(indices)
axs[0].set_xticklabels(select_features, rotation=15)
axs[0].grid(True)

# ✅ Limitar el eje Y para evitar desbordamiento
margen = (max(medias_originales) - min(medias_originales)) * 0.2
axs[0].set_ylim(min(medias_originales) - margen, max(medias_originales) + margen)

for i, val in enumerate(medias_originales):
    axs[0].annotate(f'{val:.2f}', xy=(indices[i], val),
                    xytext=(0, 10), textcoords='offset points',
                    ha='center', fontsize=9)

# Gráfico 2: Datos estandarizados
axs[1].scatter(indices, medias_zscore, color='orange', s=100, zorder=3)
axs[1].set_title('Datos Estandarizados Z-score')
axs[1].set_xlabel('Feature')
axs[1].set_ylabel('Valor Z promedio')
axs[1].set_xticks(indices)
axs[1].set_xticklabels(select_features, rotation=15)
axs[1].grid(True)

# ✅ Limitar el eje Y
axs[1].set_ylim(-0.3, 0.4)

for i, val in enumerate(medias_zscore):
    axs[1].annotate(f"z={val:.2f}", xy=(indices[i], val),
                    xytext=(0, 10), textcoords='offset points',
                    ha='center', fontsize=9)

# ✅ Anotaciones con transform para posición fija en el gráfico
axs[1].text(0.05, 0.95, f'Media global = {media:.2f}',
            transform=axs[1].transAxes, fontsize=10,
            color='darkgreen', va='top')
axs[1].text(0.05, 0.88, f'Desv. estándar = {std:.2f}',
            transform=axs[1].transAxes, fontsize=10,
            color='darkred', va='top')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, boxcox

# Variables seleccionadas
diab_sf = ['bmi', 's3', 's6']

# ── Calcular skewness para cada variable y transformación ─────────────
resultados = {}

for feat in diab_sf:
    x     = df_diab[feat].values
    x_pos = x - np.min(x) + 1e-6
    x_bc, lam = boxcox(x_pos)

    resultados[feat] = {
        'Original'      : (x,                                        None),
        'Log'           : (np.log1p(x_pos),                          None),
        'Log Reflejado' : (np.log1p(np.max(x_pos) - x_pos + 1e-6),  None),
        'Cuadrática'    : (x_pos ** 2,                               None),
        f'Box-Cox'      : (x_bc,                                     lam),
    }

transformaciones = ['Original', 'Log', 'Log Reflejado', 'Cuadrática', 'Box-Cox']
colores_col      = ['steelblue', 'mediumseagreen', 'darkorange', 'mediumpurple', 'crimson']

# ── Figura ────────────────────────────────────────────────────────────
fig, axs = plt.subplots(len(diab_sf), len(transformaciones),
                        figsize=(20, 12))
fig.suptitle('Comparativo de Transformaciones — Dataset Diabetes',
             fontsize=15, fontweight='bold', y=1.01)

for fila, feat in enumerate(diab_sf):
    mejor_sk  = min(transformaciones,
                    key=lambda t: abs(skew(resultados[feat][t][0])))

    for col, (transf, color) in enumerate(zip(transformaciones, colores_col)):
        ax   = axs[fila, col]
        datos, lam = resultados[feat][transf]
        sk   = skew(datos)

        # Resaltar la mejor transformación con borde dorado
        es_mejor = (transf == mejor_sk)
        for spine in ax.spines.values():
            spine.set_edgecolor('gold' if es_mejor else 'lightgray')
            spine.set_linewidth(3 if es_mejor else 0.8)

        ax.hist(datos, bins=20, color=color, alpha=0.75, edgecolor='white')
        ax.axvline(np.mean(datos), color='red', linestyle='--', linewidth=1.2)

        # Color del cuadro según nivel de asimetría
        color_sk = 'green' if abs(sk) < 0.5 else 'orange' if abs(sk) < 1 else 'red'
        ax.text(0.05, 0.95, f'skew={sk:.2f}',
                transform=ax.transAxes, fontsize=8, va='top',
                bbox=dict(boxstyle='round,pad=0.3',
                          facecolor=color_sk, alpha=0.35))

        # Mostrar lambda en Box-Cox
        if lam is not None:
            ax.text(0.05, 0.75, f'λ={lam:.2f}',
                    transform=ax.transAxes, fontsize=7, va='top',
                    bbox=dict(boxstyle='round,pad=0.2',
                              facecolor='white', alpha=0.6))

        # Etiquetas
        if fila == 0:
            ax.set_title(f'{transf}', fontsize=10, fontweight='bold', color=color)
        if col == 0:
            ax.set_ylabel(f'{feat}\nFrecuencia', fontsize=9, fontweight='bold')
        else:
            ax.set_ylabel('')
        ax.set_xlabel('Valor', fontsize=7)
        ax.tick_params(labelsize=7)
        ax.grid(True, alpha=0.25)

plt.tight_layout()
plt.show()

# ── Tabla resumen de skewness ─────────────────────────────────────────
print("\n📊 Tabla Comparativa de Skewness")
print(f"  {'':15}", end='')
for t in transformaciones:
    print(f"  {t:>14}", end='')
print()
print("  " + "-" * 87)

for feat in diab_sf:
    print(f"  {feat:<15}", end='')
    skews = [skew(resultados[feat][t][0]) for t in transformaciones]
    mejor = min(range(len(skews)), key=lambda k: abs(skews[k]))
    for j, sk in enumerate(skews):
        estrella = ' 🏆' if j == mejor else '   '
        print(f"  {sk:>+10.4f}{estrella}", end='')
    print()

print("\n  🟢 |skew| < 0.5  Simétrica")
print("  🟠 |skew| < 1.0  Leve asimetría")
print("  🔴 |skew| ≥ 1.0  Alta asimetría")
print("  🏆 Mejor transformación por variable")
```
