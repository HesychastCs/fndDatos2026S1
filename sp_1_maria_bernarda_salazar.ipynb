## <span style="color:Blue;"><strong>Preparación de la fuente de datos y librerías</strong></span>

```python
# Librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo general
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Datos
# URL directa al archivo ZIP que contiene los datos
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00320/student.zip"

# Descarga y extracción
import zipfile, io, requests

r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall("data/student_data")

# Crear dataset
df = pd.read_csv("data/student_data/student-mat.csv", sep=";")
```

## <span style="color:Blue;"><strong>4. Visualizaciones exploratorias básicas</strong></span>

<div style="background-color:#FCE4EC; padding:10px; border-radius:8px;">
<b>- Gráficos de dispersión para relaciones entre variables.</b><br>
<b>- Breve interpretación de lo que muestran las gráficas.</b><br>
<b>- Histogramas o gráficos de barras para variables numéricas y categóricas.</b><br>

```python
# Histograma: distribución de edad
sns.histplot(df['age'], bins=6, kde=True, color='skyblue')
plt.title("Distribución de edad de los estudiantes")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Gráfico de barras: tiempo de estudio
# Mapeo de categorías
studytime_labels = {
    1: "<2 h/sem",
    2: "2–5 h/sem",
    3: "5–10 h/sem",
    4: ">10 h/sem"
}

# Columna con etiquetas
df['studytime_label'] = df['studytime'].map(studytime_labels)

# Gráfico 
sns.countplot(x='studytime_label', data=df, hue='studytime_label', palette='viridis',legend=False)
plt.title("Tiempo de estudio semanal")
plt.xlabel("Categoría de tiempo de estudio")
plt.ylabel("Número de estudiantes")
plt.show()
```

```python
# Gráfico de barras: género
sns.countplot(x='sex', data=df,hue='sex', palette='pastel',legend=False)
plt.title("Distribución por género")
plt.xlabel("Género")
plt.ylabel("Número de estudiantes")
plt.show()
```

```python
# Gráfico de dispersión: calificación final vs ausencias
sns.scatterplot(x='absences', y='G3', hue='sex', data=df)
plt.title("Relación entre ausencias y calificación final (G3)")
plt.xlabel("Número de ausencias")
plt.ylabel("Calificación final (G3)")
plt.show()
```

```python
# Gráfico de dispersión: tiempo de estudio vs calificación final
plt.figure(figsize=(8,6))
sns.boxplot(x='studytime_label',y='G3',hue='studytime_label',data=df,palette='coolwarm',width=0.5,fliersize=0,legend=False);
sns.stripplot(x='studytime_label',y='G3',data=df,color='black',alpha=0.5,jitter=True);

plt.title("Calificación final según tiempo de estudio");
plt.xlabel("Tiempo de estudio");
plt.ylabel("Calificación final (G3)");
plt.grid(axis='y',linestyle='--',alpha=0.7);
plt.show()
```

```python
# Gráfico de dispersión con líneas de tendencia
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# G1 vs G2
sns.regplot(x='G1', y='G2', data=df, ax=axes[0],
            scatter_kws={'s': 50, 'alpha': 0.7},
            line_kws={'color': 'red'})
axes[0].set_title('Relación entre G1 y G2', fontsize=14)
axes[0].set_xlabel('Nota Parcial 1 (G1)', fontsize=12)
axes[0].set_ylabel('Nota Parcial 2 (G2)', fontsize=12)

# G2 vs G3
sns.regplot(x='G2', y='G3', data=df, ax=axes[1],
            scatter_kws={'s': 50, 'alpha': 0.7},
            line_kws={'color': 'blue'})
axes[1].set_title('Relación entre G2 y G3', fontsize=14)
axes[1].set_xlabel('Nota Parcial 2 (G2)', fontsize=12)
axes[1].set_ylabel('Nota Final (G3)', fontsize=12)

plt.tight_layout()
plt.show()
```
